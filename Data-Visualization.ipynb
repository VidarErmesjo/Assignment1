```python
import pandas as panda
```

### Dataset:

```python
data = panda.read_csv("Ruter_data.csv", skiprows=0, delimiter=';', names=None)
data.head()
```

### Scatter plot:

```python
data.plot.scatter(x='Dato', y='Passasjerer_Ombord', title='Scatter plot')
```

### Line plot:

```python
data.drop(['TurId', 'Fylke', 'Område', 'Kommune', 'Holdeplass_Fra', 'Holdeplass_Til', 'Linjetype', 'Linjefylke', 'Linjenavn', 'Linjeretning'], axis=1).plot.line(title="Line plot")
```

### Histogram:

```python
data['Passasjerer_Ombord'].plot.hist()
```

```python
data.plot.hist(subplots=True, layout=(2,2), figsize=(10, 10), bins=20)
```
